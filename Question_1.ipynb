```python
import matplotlib.pyplot as plt
from matplotlib import interactive

file = open("enel420_grp_19.txt", "r") #Opening data file
data = []

for line in file:                      #Seperating and appending data points to an array
    lineOfPoints = line.split('  ')
    for point in lineOfPoints:
        data.append(float(point)) 

t = list(range(0, len(data)))          #Creating time array
t[:] = [x/1024 for x in t]

#Plotting and other oddities
plt.rcParams["figure.figsize"] = (15, 10)
plt.rcParams["figure.dpi"] = 150
fig = plt.figure()
fig.suptitle('ECG Signal, Question 1', fontsize=20)
plt.xlabel('Time (s)', fontsize=18)
plt.ylabel('Voltage (μV)', fontsize=16)
fig.savefig('Question 1.jpg')
plt.plot(t, data)
```
