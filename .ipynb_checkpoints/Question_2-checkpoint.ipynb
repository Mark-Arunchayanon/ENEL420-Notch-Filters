```python
# %matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib import interactive
import numpy as np
from scipy import fftpack
import plotly.express as px
```

```python
fs = 1024 #sampling frequency 
```

```python
file = open("enel420_grp_19.txt", "r")
data = []

for line in file:
    lineOfPoints = line.split('  ')
    for point in lineOfPoints:
        data.append(float(point))

t = list(range(0, len(data)))
t[:] = [x/1024 for x in t]


#plot from question 1

#Plotting and other oddities
plt.rcParams["figure.figsize"] = (15, 10)
plt.rcParams["figure.dpi"] = 150
fig = plt.figure()
fig.suptitle('ECG Signal, Question 1', fontsize=20)
plt.xlabel('Time (s)', fontsize=18)
plt.ylabel('Voltage (V)', fontsize=16)
fig.savefig('Question 2.jpg')
plt.ion()
plt.plot(t, data)
plt.show()
```

```python
figure = px.line(x=t, y=data, labels = {'x':'Time (s)', 'y':'Voltage (V)'}, title='Raw Data (Q1)', width=750)
figure.show()
```

```python
plt.magnitude_spectrum(data, Fs=fs)
```

```python
dataFFT = fftpack.fft(data)
power = np.abs(dataFFT)

sampleFreq = fftpack.fftfreq(len(data), d=(1/fs))
```

```python
plt.figure(figsize = (15, 10))
plt.plot(sampleFreq, power)
```

```python
# Find the peak frequency: we can focus on only the positive frequencies
pos_mask = np.where(sampleFreq > 0)
freqs = sampleFreq[pos_mask]
peak_freq = freqs[power[pos_mask].argmax()]
```
